# SMILES

Simplified Molecular Input Line Entry System


Es una especificación en forma de notación lineal para describir la estructura de especies químicas mediante cadenas ASCII cortas. La mayoría de los editores de moléculas pueden importar cadenas SMILES para convertirlas en dibujos bidimensionales o modelos tridimensionales de las moléculas.

La especificación SMILES original se inició en la década de 1980. Desde entonces, ha sido modificada y ampliada. En 2007, se desarrolló un estándar abierto llamado OpenSMILES en la comunidad química de código abierto.

```python
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors

antiTBC = ['c1(C(=O)N)nccnc1','C1=CN=CC=C1C(=O)NN','CCC(CO)NCCNC(CC)CO']
nombres = ["pirazinamida", "isoniazida", "etambutol"]

mol_lista = []
lista_de_etiquetas = []

for i, smi in enumerate(antiTBC):
    mol = Chem.MolFromSmiles(smi)
    mol_lista.append(mol)

    logp = Descriptors.MolLogP(mol)
    lista_de_etiquetas.append(f"{nombres[i]} | logP = {logp:.2f}")

Draw.MolsToGridImage(mol_lista,legends=lista_de_etiquetas)
```

# SMARTS

SMILES Arbitrary Target Specification

Es un potente lenguaje químico, una extensión de SMILES, que se utiliza para definir y buscar subestructuras o patrones moleculares específicos. Actúa como "expresiones regulares" en química y permite realizar consultas precisas sobre tipos de átomos, enlaces, cargas, sistemas de anillos y disposiciones complejas. Permite a los químicos encontrar moléculas con grupos funcionales, estructuras o propiedades electrónicas particulares, cruciales para el descubrimiento de fármacos y la quimioinformática.

**Conceptos clave**:

Búsqueda de subestructuras: SMARTS describe subestructuras (p. ej., cualquier amina, cualquier anillo halogenado) en lugar de moléculas individuales, lo que lo hace ideal para búsquedas extensas.

**Sintaxis**:

Utiliza la notación SMILES habitual (átomos, enlaces como -, =, #), pero añade operadores lógicos (& para AND, , para OR, ! para NOT) y símbolos especiales para propiedades como carga ([+1]), recuento de hidrógeno (H₂), pertenencia al anillo (R) y quiralidad (@). Expresividad: Permite especificaciones detalladas, como [#6]=,:[#6] (carbono con doble enlace o aromático a otro carbono) o C(c)(c)C([Br,Cl,F])([Br,Cl,F])([Br,Cl,F]) (un diaril etano con tres halógenos).
Aplicaciones: Se utiliza en bases de datos (como ZINC), software (OpenEye, Daylight) y herramientas (SMARTS.plus) para tareas como el filtrado de compuestos de plomo, la definición de entornos químicos (p. ej., para toxicidad) y el análisis de tipos de reacción.

**Ejemplos de sintaxis SMARTS**:

[c]: Cualquier carbono aromático.
[#8]: Cualquier átomo de oxígeno.
[F,Cl,Br,I]: Cualquiera de los cuatro primeros halógenos.
[!c]: Cualquier carbono no aromático.
[C;H3]: Un carbono con exactamente tres hidrógenos (un grupo metilo). C-C: Un carbono alifático unido mediante un enlace simple a otro.
En esencia, SMARTS proporciona un lenguaje versátil para consultar el universo químico basándose en patrones estructurales, lo que facilita el análisis químico complejo y la recuperación de información.

En muchas aplicaciones comunes, como el procesamiento de textos, necesitamos buscar una cadena de texto específica. En quimioinformática, nos encontramos con situaciones similares en las que queremos determinar si los átomos de una molécula coinciden con un patrón específico. Existen varios casos de uso donde esto puede ocurrir:

• Búsqueda en una base de datos de moléculas para identificar moléculas que contienen una subestructura específica

• Alineación de un conjunto de moléculas en una subestructura común para mejorar la visualización

• Resaltado de una subestructura en un gráfico

• Restringir una subestructura durante un cálculo

```python
from rdkit import Chem
from rdkit.Chem.Draw import MolsToGridImage
smiles_list = ["CCCOCCO","CCOCC","CCNCC","CCSCC"]
mol_list = [Chem.MolFromSmiles(x) for x in smiles_list]
MolsToGridImage(mol_list, molsPerRow=4)
```

```python
query = Chem.MolFromSmarts("CCC")
match_list = [mol.GetSubstructMatch(query) for mol in mol_list]
MolsToGridImage(mols=mol_list, molsPerRow=4,highlightAtomLists=match_list)
```

```python
query = Chem.MolFromSmarts("C*C")
match_list = [mol.GetSubstructMatch(query) for mol in mol_list]
MolsToGridImage(mols=mol_list, molsPerRow=4,highlightAtomLists=match_list)
```

```python
query = Chem.MolFromSmarts("O")
match_list = [mol.GetSubstructMatch(query) for mol in mol_list]
MolsToGridImage(mols=mol_list, molsPerRow=4,highlightAtomLists=match_list)
```
