## Interpretación de los términos energéticos en AutoDock Vina 1.2.7

En **AutoDock Vina 1.2.7**, la función de *scoring* descompone la energía de unión estimada en varios términos que permiten analizar tanto la calidad del acoplamiento como el origen físico de la afinidad observada. A continuación se describe el significado de cada uno de los valores reportados por el método `v.energies()`.

---

### 1. Afinidad (`aff`)
- Corresponde a la **energía libre de unión estimada** (*ΔG<sub>binding</sub>*), expresada en kcal/mol.
- Es el **parámetro principal utilizado para el ranking de ligandos**.
- Valores más negativos indican una afinidad de unión más favorable.
- Se calcula aproximadamente como:

$$
\Delta G \approx \mathrm{Inter} + \mathrm{Tors}
$$

Ejemplo (Pose 1):

$$
-4.73 + 0.26 = -4.47\ \mathrm{kcal/mol}
$$

---

### 2. Energía intermolecular (`inter`)
- Representa la **energía de interacción ligando–receptor**.
- Incluye contribuciones de:
  - Interacciones hidrofóbicas  
  - Puentes de hidrógeno  
  - Interacciones π–π  
  - Repulsión estérica  
- Es el **principal término favorable** que estabiliza el complejo.

---

### 3. Energía interna (`internal`)
- Corresponde a la **energía interna del ligando** en la conformación adoptada durante el docking.
- Penaliza conformaciones geométricamente desfavorables.
- Valores cercanos a cero indican que el ligando **no está significativamente forzado**.
- No contribuye directamente al cálculo final de la afinidad, pero influye en el muestreo conformacional.

---

### 4. Penalización torsional (`tors`)
- Término entrópico asociado a la **pérdida de libertad conformacional** del ligando al unirse al receptor.
- Depende del número de enlaces rotables.
- Es **siempre positivo** (desfavorable).
- Ligandos más flexibles presentan valores de `tors` más altos.

---

### 5. Energía del estado no unido (`unbound`)
- Energía de referencia del ligando en el **estado no unido**.
- Se emplea para normalización interna del modelo.
- No se utiliza directamente para comparar afinidades entre ligandos.
- Valores cercanos a cero son normales y esperados.

---

## Resumen de los términos energéticos

| Término | Descripción | Rol en el docking |
|-------|------------|------------------|
| `aff` | Energía libre de unión estimada (ΔG) | Ranking principal |
| `inter` | Interacciones ligando–receptor | Contribución favorable |
| `internal` | Energía interna del ligando | Calidad conformacional |
| `tors` | Penalización entrópica | Contribución desfavorable |
| `unbound` | Estado de referencia | Normalización |

---

```python
from vina import Vina

v = Vina()

for file in ligand_files:
    v.set_receptor("mpro.pdbqt")
    v.set_ligand_from_file(file)
    v.compute_vina_maps([-22.194, 18.772, -24.459], [30, 30, 30])

    v.dock(
        exhaustiveness=4,
        n_poses=3
    )

    energias = v.energies()

    print(f"\nLigando: {file}")
    for i, e in enumerate(energias, start=1):
        print(f"Pose {i}: Afinidad = {e[0]:.2f} kcal/mol")

    v.write_poses(file.replace(".pdbqt", "-resultados.pdbqt"))
```
